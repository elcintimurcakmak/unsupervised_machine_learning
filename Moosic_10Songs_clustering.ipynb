```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/content/2_spotify_10_songs.csv')
df
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
selected_feas=['danceability', 'energy', 'acousticness', 'tempo', 'valence']
X=df[selected_feas]
X
```

```python
kmeans=KMeans(n_clusters=3, random_state=123)
kmeans.fit(X)
```

```python
X['clusters']=kmeans.labels_
X
```

```python
X.sort_values(by='clusters')
```

```python
X.index = df['song_name']
```

```python
plt.scatter(x=X.iloc[:,0],
            y=X.iloc[:,1],
            c=X['clusters'],
            cmap='viridis')

plt.scatter(x=kmeans.cluster_centers_[:,0],
            y=kmeans.cluster_centers_[:,1],
            c='red',
            marker='X',
            s=100,
            label='Centroids')

for idx, row in X.iterrows():
    plt.annotate(idx, (row['danceability'], row['energy']), xytext=(5,0), textcoords='offset points')

plt.title('KMeans for 10 Spotify Songs')
plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.legend()
plt.show()
```
